```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['buying']=le.fit_transform(df['buying'])
```

```python
df['lug_boot']=le.fit_transform(df['lug_boot'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf=KNeighborsClassifier(n_neighbors=4,p=1,metric= "minkowski")
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred_train=clf.predict(X_train)
y_pred_test=clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_train,y_pred_train)
```

```python
metrics.confusion_matrix(y_test,y_pred_test)
```

```python
print("Train Data")
print(metrics.classification_report(y_train,y_pred_train))
print("Test Data")
print(metrics.classification_report(y_test,y_pred_test))
```

```python
f1_score=metrics.f1_score(y_train,y_pred_train,average='weighted')
accuaracy=metrics.accuracy_score(y_train,y_pred_train)
```

```python
f1_score=metrics.f1_score(y_test,y_pred_test,average='weighted')
accuaracy=metrics.accuracy_score(y_test,y_pred_test)
```
